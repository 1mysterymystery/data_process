```python
import pandas as pd
```

```python
df = pd.read_csv(r'/usr/share/sc5tadm/gotoHomePubData/dataset/ConutiousReforming/山东润泽_多位号/v5_baseline/SDRZ_v5_baseline.csv')
```

```python
import json
# 读取JSON文件
with open(r'/usr/share/sc5tadm/gotoHomePubData/dataset/ConutiousReforming/山东润泽_多位号/v5_baseline/SDRZ_del_v5_baseline_continous.json', 'r') as f:
    segments = json.load(f)

# 创建一个空列表来存储需要保留的行索引
keep_rows = []

# 遍历JSON文件中的每个片段
for segment, rows in segments.items():
    start, end = rows
    length = end - start + 1
    
    # 如果长度大于等于400,则将这个片段的行索引添加到keep_rows列表中
    if length >= 300:
        keep_rows.extend(range(start, end))

# 根据keep_rows列表中的索引过滤出需要保留的行
filtered_df = df.iloc[keep_rows]

# 保存过滤后的数据到新的CSV文件/usr/share/sc5tadm/gotoHomePubData/dataset/ConutiousReforming/山东润泽/处理后数据v1/SDRZ_del_json_v1.csv
filtered_df.to_csv('/usr/share/sc5tadm/gotoHomePubData/dataset/ConutiousReforming/山东润泽_多位号/v5_baseline/SDRZ_v51_baseline.csv', index=False)
```
